```python
dizi1 = np.linspace(0,10,20)
```

```python
dizi2
```

```python
plt.plot(dizi1,dizi2,"g--")
```

```python
plt.subplot(1,2,1)
plt.plot(dizi1,dizi2,"r*-")
plt.subplot(1,2,2)
plt.plot(dizi2,dizi1,"g--")
```

```python
figure = plt.figure()
figureAxes = figure.add_axes([0.2,0.2,0.9,0.9]) ## argümanları liste içerisinde vermem gerekiyo
figureAxes.plot(dizi1,dizi2,"g")
figureAxes.set_xlabel("X ekseni ")
figureAxes.set_ylabel("y ekseni")
figureAxes.set_title("grafik başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.1,0.3,0.3,0.3])

eksen1.plot(dizi1,dizi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("grafik başlığı")

eksen2.plot(dizi1,dizi2,"g")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("grafik başlığı")

```

```python
plt.subplots()
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows = 1 , ncols=2)
type(benimEksen)

for eksen in benimEksen:
    eksen.plot(dizi1,dizi2)
    eksen.set_xlabel("x ekseni")
```

```python
yeniFigur = plt.figure(dpi = 150)
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(dizi1,dizi1**2,label = "numpy dizisi **2")
yeniEksen.plot(dizi1,dizi1**3,label = "numpy dizisi **3")
yeniEksen.legend(loc = 3)
```
